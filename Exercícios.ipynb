```python
def swapColumns(col1, col2, df):
    # Armazenando as colunas do df em uma variável
    cols = df.columns
    
    # 'Pegando'(get), os índices das colunas para a troca
    col_1 = cols.get_loc(col1)
    col_2 = cols.get_loc(col2)
    
    # trocando as duas colunas usando o índice
    df[[cols[col_1], cols[col_2]]] = df[[cols[col_2], cols[col_1]]]
    
    # Renomeando as colunas para efetivar a troca, inplace para alterar no df original
    df.rename(columns = {col1:col2, col2:col1}, inplace = True)
```

```python
swapColumns('a', 'b', df)
```

```python
def inter(x,y,df):
    lis=list(df.columns) # Criando uma lista com as colunas
    index_x=lis.index(x) # Selecionando uma coluna para ser armazenada na variável x
    index_y=lis.index(y) # Selecionando uma coluna para ser armazenada na variável y
    lis[index_x]=y # Inverter as variáveis para fazer a troca (index_x para y)
    lis[index_y]=x # Inverter as variáveis para fazer a troca (index_y para x)
    df=df[lis]
    return df

a=inter('a','b',df) # Digitar as colunas para serem trocadas
a
```

```python
lin=[]
dist=[]
for i in range(len(aux)):
    dif_p =(aux[i,:]-aux)# distancia entre ponto da linha i x todas as linhas (10x4)
    som_prod=(dif_p@(dif_p.T))# calc soma do produto de cada linha utilizando produto de matrizes: (10x4)@(4x10)-> 10x10
    mod= som_prod.diagonal()**(0.5) # diagonal igual ao modulo ao quadrado de cada linha
    mod = np.delete(mod,[i]) # retirando o valor 0 que é a dinstancia de um ponto pra ele mesmo
    lin.append(np.argmin(mod))
```
